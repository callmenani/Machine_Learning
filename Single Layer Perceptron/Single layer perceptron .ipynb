```python
import numpy as np
```

```python
def step_function(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python
def logical_NOT(x):
    w = np.array([-1]) 
    b = 0
    x = np.array(x) 
    z = np.dot(w, x) + b 
    y = step_function(z)
    return y
```

```python
def logical_NAND(x):
    w = np.array([-1, -1]) 
    b = 2 
    x = np.array(x) 
    z = np.dot(w, x) + b 
    y = step_function(z) 
    return y
```

```python
def logical_NOR(x):
    w = np.array([-1, -1])
    b = 0 
    x = np.array(x) 
    z = np.dot(w, x) + b
    y = step_function(z) 
    return y
```

```python
print(logical_NOT(0))
```

```python
print(logical_NOT(1))
```

```python
print(logical_NAND([0, 0])) 
print(logical_NAND([0, 1])) 
print(logical_NAND([1, 0])) 
print(logical_NAND([1, 1]))
```

```python
print(logical_NOR([0, 0])) 
print(logical_NOR([0, 1])) 
print(logical_NOR([1, 0])) 
print(logical_NOR([1, 1])) 
```
