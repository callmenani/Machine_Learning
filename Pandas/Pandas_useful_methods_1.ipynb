```python
def rate(price):
   if price<10:
     return '$'
   elif price>=10 and price<30:
     return '&&'
   else:
     return '$$$'
```

```python
df['rate']=df['total_bill'].apply(rate)
```

```python
def quality(t_bill,tip):
  if tip/t_bill > 0.25:
    return 'generous'
  else : 
    return 'other'
```

```python
df['Quality']=df[['total_bill','tip']].apply(lambda df : quality(df['total_bill'],df['tip']),axis=1)
```

```python
df.head(3)
```

```python
# this is how we use apply() to multiple columns
```

```python
df['Quality']=np.vectorize(quality)(df['total_bill'],df['tip'])
```
